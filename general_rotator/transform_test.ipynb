```python
import numpy as np

A = np.asarray([[1,0,0],[0,2,0],[0,0,3]])
x = np.asarray([1,1,1])

y = A@x
B = x@A

print(y)
print(B)
```

```python
def Bfield(x_):
    return np.asarray([x_[0],2*x_[1],3*x_[2]])

def transform(R,vector):
    def vector_transformed(x_):
        return vector(R@x_)
    return vector_transformed

def transform_sel(R,vector,selection=(0,1,2)):
    # print([idx for idx in selection])
    def vector_transformed(x_):
        return np.asarray([(vector(R@x_))[idx] for idx in selection])
    vector_transformed.attr = 'xyz'
    return vector_transformed

def transformf_sel(R,vector,selection=(0,1,2)):
    # print([idx for idx in selection])
    def vector_transformed(x_):
        return np.asarray([(np.transpose(R)@vector(R@x_))[idx] for idx in selection])
    return vector_transformed

def transformall_sel(R,offset,vector,selection=(0,1,2)):
    # print([idx for idx in selection])
    def vector_transformed(x_):
        return np.asarray([(np.transpose(R)@vector(R@x_ + offset))[idx] for idx in selection])
    return vector_transformed

myfunct = transform_sel(A,Bfield)
print(Bfield(x))
print(transform(A,Bfield)(x))
print(transform_sel(A,Bfield,selection=(1,2))(x))
print(transformf_sel(A,Bfield,selection=(1,2))(x))
print(myfunct.attr)
    
```

```python
def transform_vector_field(R,offset,vector,selection=(0,1,2)):
    def vector_transformed(x_):
        return np.asarray([(R@vector((np.transpose(R)@x_) + offset))[idx] for idx in selection])
    return vector_transformed
```
